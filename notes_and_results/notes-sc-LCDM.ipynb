```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
def Omega(z, Om0):
    E = Om0 * (1+z)**3 + (1-Om0)
    return Om0 * (1+z)**3 / E
```

```python
def Del_vir(Om):
    x = Om - 1
    return 18*np.pi**2 + 82*x - 39*x**2
```

```python
z = np.logspace(-3, 2)
plt.plot(Om, Del_vir(Omega(z, 0.3)))
plt.plot(Om, Del_vir(Omega(z, 1)))
plt.xscale('log')
```

```python
Om = np.linspace(0,1,50)
```

```python
plt.plot(Om, Del_vir(Om))
plt.plot(Om, Del_vir(Om/Om))
```
